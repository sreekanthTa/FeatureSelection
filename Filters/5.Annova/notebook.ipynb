```python
from scipy.stats import stats
# Given Data
data = {
    "Education": ["Bachelor's", "Bachelor's", "Bachelor's", "Master's", "Master's", "Master's", "PhD", "PhD", "PhD"],
    "Salary": [50000, 55000, 53000, 60000, 62000, 65000, 80000, 82000, 85000]
}

# Convert to numpy arrays
bachelors = np.array([50000, 55000, 53000])
masters = np.array([60000, 62000, 65000])
phd = np.array([80000, 82000, 85000])
```

```python
# Perform ANOVA
f_stat, p_value = stats.f_oneway(bachelors, masters, phd)

print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.5f}")
```

```python
# Decision Rule
alpha = 0.5

if p_value > alpha:
    print("Reject the NULL hypthosis, Education Level Effects Salary")
else:
    print("Fail to Reject the NULL Hypthesis: No Significant effect")
```
