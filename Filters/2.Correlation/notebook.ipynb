```python
# Sample data
height = [150, 160, 170, 180, 190]  # in cm
weight = [50, 60, 70, 80, 90]  # in kg
income = [30000, 40000, 50000, 60000, 70000]  # in dollars

data = {
    "Age": [22, 25, 47, 52, 46, 56, 60, 62, 30, 40],
    "Salary": [25000, 32000, 60000, 72000, 58000, 80000, 85000, 90000, 40000, 50000],
    "Experience": [1, 3, 15, 20, 14, 25, 30, 35, 5, 10]
}
df = pd.DataFrame(data)

# Compute covariance matrix
cov_matrix = df.cov()
print("Covariance Matrix:\n", cov_matrix)

# Compute correlation matrix
corr_matrix = df.corr()
print("\nCorrelation Matrix:\n", corr_matrix)
```

```python

# Compute correlation matrix
corr_matrix = df.corr().abs()  # Absolute correlation values
threshold = 0.9  # Define the threshold

# Find highly correlated feature pairs
high_corr_features = set()

# Iterate through correlation matrix to find pairs
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] > threshold:
            high_corr_features.add(corr_matrix.columns[i])

# Drop highly correlated features
df_filtered = df.drop(columns=high_corr_features)
print("Highly Correlated Features are ", high_corr_features)
print("\nFiltered Data (After Removing Highly Correlated Features):\n", df_filtered)
```
