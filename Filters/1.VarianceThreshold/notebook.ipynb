```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold

data = {
    "Feature1": [1, 1, 1, 1, 1],  # No variance (constant)
    "Feature2": [0, 1, 0, 1, 1],  # Some variance
    "Feature3": [2, 2, 2, 2, 2],  # No variance (constant)
    "Feature4": [10, 11, 12, 13, 14]  # High variance
}

df = pd.DataFrame(data)

df
```

```python
selector = VarianceThreshold(threshold=0.1)

df_selected = selector.fit_transform(df)

selected_features = df.columns[selector.get_support()]

print("selected features are", selected_features)

print("\nFiltered Data:\n", pd.DataFrame(df_selected, columns=selected_features))
```
