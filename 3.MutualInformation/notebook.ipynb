```python
np.random.seed(42)  # Set seed for reproducibility

# Features (2D Array: Age and Salary)
X = np.array([
    [10, 1000],  # Age 10, Salary 1000
    [15, 2000],  # Age 15, Salary 2000
    [20, 3000],  # Age 20, Salary 3000
    [25, 4000],  # Age 25, Salary 4000
    [30, 5000],  # Age 30, Salary 5000
    [35, 6000],  # Age 35, Salary 6000
    [40, 7000],  # Age 40, Salary 7000
    [45, 8000],  # Age 45, Salary 8000
    [50, 9000],  # Age 50, Salary 9000
    [55, 10000]  # Age 55, Salary 10000
])

Y = np.array(['Low', 'Low', 'Medium', 'Medium', 'High', 'High', 'Very High', 'Very High', 'Low', 'Low'])
```

```python
label_encoder = LabelEncoder()
y_encoded=label_encoder.fit_transform(Y)
y_encoded
```

```python
mi_class = mutual_info_classif(X, y_encoded)
mi_class
```

```python
# Convert to DataFrame for better visualization
df_mi = pd.DataFrame({
    'Feature': ['Age', 'Salary'],
    'MI Score': mi_class
})


df_mi
```
